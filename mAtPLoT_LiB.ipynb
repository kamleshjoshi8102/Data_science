```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
type(x)
```

```python
y=2*x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line_plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='black' , linestyle=':' , linewidth=5)
plt.show()
```

```python
plt.plot(x,y,color='r' , linestyle=':' , linewidth=5)
plt.show()
```

```python
y1=x*3
y2=4*x
```

```python
plt.plot(x,y,y1,y2)
plt.title("Line_plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y1,color='r',linestyle=":" , linewidth=6)
plt.plot(x,y2,color='g',linestyle="-" , linewidth=4)
plt.title("LINe Plot")
plt.grid(True)
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.subplot(1,2,1) #(row,column , in which cell)
plt.subplot(1,2,2)
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.title("Sine plot")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x,np.sin(x))
plt.title("Sine plot")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1 , 2 , 1)
plt.plot(x,y1,color='r' , linestyle=":" , linewidth=6)
plt.grid(True)

plt.subplot(1 , 2 , 2)
plt.plot(x,y1,color='g' , linestyle="--" , linewidth=4)
plt.grid(True)

plt.show()
```

```python
values=list((student.values()))
values
```

```python
#bar plot
```

```python
plt.barh(names , values , color='r')
plt.title("Horizontal")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.bar(names , values , color='black')
plt.title("Horizontal")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
